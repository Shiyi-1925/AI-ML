# Reminder

<p style="text-align:justify";> Connected operators  have been introduced by Serra and Salembier in [Salembier95flatzones]. These operators do not modify individual pixels, but act on connected components (*CC*), for binary images, or on flat-zones (for gray level images). According to [Salembier95flatzones], a *CC*  of a set is a set of points that are connected by a path included in the set, while flat-zones (*FZ*) of an image *I* are the largest *CC*s where the image is constant. We denote *FZ(I)* the set of flat-zones of image *I*.</p>

Connected operators ($ \Psi $) can only merge *FZ* but do not break them into pieces. As a consequence, connected operators cannot add new boundaries, nor shift existing contours, but only simplify the image, removing contours. In other words, each *FZ* of the original image should be included in a single *FZ* of the resulting image: $ \forall B \in FZ(I), \exists 
C \in FZ(\Psi (I))~ /~ B \subset C$.

<img src="../illustrations/attributes_illustration.png" align="center" width="400px;" title="Attributes illustration"/> 

Attribute openings (*AO*) are connected operators. For binary images, each *CC* of the original image is analyzed and an attribute is computed on it. Examples of attributes are the *CC* area, height or width of the bounding box or any other measure we can define on a given *CC* such as circularity (other examples can be found in [Breen96]). An increasing criterion *T* is defined based on this attribute, usually in the form of a threshold on the attribute. Criterion *T* is said to be increasing if the fact that a set *C* satisfies *T*, implies that any other set *B* containing *C* $(B \supseteq C)$ satisfies *T* too. Then the *AO* of *I* is equal to the union of *CC* of *I* verifying criterion *T*. For example, an area opening of size $\lambda$ ($\gamma_{\lambda}^{area}(I)$) is equal to the union of *CC*s of *I* which area is greater than or equal to $\lambda$.

# Area opening

The code beolw produces a gray-scale image and two area openings, of sizes *100* and *500* pixels respectively. The regional maxima of these images are represented in white. We can see that while the original image has many small maxima, the filtered images have larger and fewer maxima. The remaining maxima have an area of at least *100* and *500* pixels respectively, according to the corresponding opening size.

You can try this operator with other images or other sizes hereafter.

```python
import sys
sys.path.append('/home/commun/tp-morpho/')
from tp_init import *
```

```python
nl = HexSE()
im = Image(images_dir+"Bin/metal1.png")
im2 =Image(im)
disp(im, figsize=(5, 5))
print("Original image")

for filter_size in [100, 250, 500]:
    areaOpen(im,filter_size,im2,nl)
    disp(im2, figsize=(5, 5))
    print("Area opening of size ", filter_size)
```

```python
nl = HexSE()
#im = Image(images_dir+"Gray/foreman.png")
#im = Image(images_dir+"Gray/tools.png")
im = Image(images_dir+"WeizmannDataBase/yokohm060409_dyjsn191.png") #110016671724.png
im2, imtmp = Image(im), Image(im)

maxima(im,imtmp,nl)
imOpenMax = colorHighlight(im, imtmp)
disp([im,imOpenMax])
print ("Original image and its regional maxima")

# Apply area open of size 100 in this example
for filter_size in [100,500]:
    areaOpen(im,filter_size,im2,nl)

    # Verify the maxima of the filtered image
    maxima(im2,imtmp,nl)
    imOpenMax = colorHighlight(im2, imtmp)
    #compare(imtmp,">",0,255,im,imtmp)

    # Display the result
    disp([im2,imOpenMax])
    print ("areaOpen size", filter_size, " and its regional maxima")
```
